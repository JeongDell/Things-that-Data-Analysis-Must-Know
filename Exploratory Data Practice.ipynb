```python
# 필요한 패키지 임포트

import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

sns.set(color_codes=True)
%matplotlib inline
```

```python
#데이터 불러오기
df=pd.read_csv('hotel_bookings.csv')

#데이터 샘플 확인(head: index=0에서 시작해서 5개 불러오기)
df.head()
```

```python
#각 컬럼의 속성 및 결측치 확인
df.info()
```

```python
#각 컬럼의 통계치 확인, mean: 평균, std: 표준편차

df.describe()
```

```python
#각 칼럼의 왜도 확인
#skew: numeric(수치형 데이터만 처리 가능)

df.skew(numeric_only=True)
```

```python
#각 칼럼의 척도 확인, numeric만 처리 가능

df.kurtosis(numeric_only=True)
```

```python
#특정 변수 분포 시각화
#기본적인 y축은 어떻게 설정되는가?

sns.displot(df['lead_time'])
```

```python
df['lead_time']
```

```python
#호텔 구분에 따른 load_time 분포 차이 시각화

#violinpolt: 가장 큰 값과 작은 값을 모두 하나로 연결하여 하나의 도형으로 표시
#stripplot: 데이터 포인트가 중복되어 범주별 분포를 그림

sns.violinplot(x='hotel', y='lead_time', data=df,inner=None, color='.8')
sns.stripplot(x='hotel', y='lead_time', data=df, size=1)
```
