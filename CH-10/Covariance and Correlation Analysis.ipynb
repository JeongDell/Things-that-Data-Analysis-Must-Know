```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

```

```python
#산점도 행렬 시가화

sns.set(font_scale=1.1) ##폰트 크기 설정
sns.set_style('ticks') ##축 눈금 설정
sns.pairplot(df,
             diag_kind='kde' # 상관계수가 1이면 분포로 표시
             )

plt.show
```

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# 매트릭스의 상단 삼각행렬을 True, 하단 삼각행렬을 False로 변환
#np.ones_like(df.corr())는 df.corr()와 동일한 형태의 배열을 생성, 모든 요소가 1
#np.triu 함수는 이 배열의 상삼각행렬을 반환. 대각선 아래의 모든 요소를 0으로 설정, 대각선 위의 요소는 1로 유지
#dtype=bool을 사용하여 True/False 마스크 배열을 
# 상단 삼각행렬을 True로 설정하여 마스크 배열을 생성. 이 마스크는 히트맵에서 대각선 아래의 값들을 숨기는 데 사용

mask = np.triu(np.ones_like(df.corr(), dtype=bool))

# 히트맵 그래프 생성
fig, ax = plt.subplots(figsize=(15, 10))
sns.heatmap(df.corr(),
            mask=mask,
            vmin=-1,
            vmax=1,
            annot=True,
            cmap='RdYlBu_r',
            cbar=True)

ax.set_title('Wine Quality Correlation', pad=15)
plt.show()

```
